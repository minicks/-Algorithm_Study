```python
INPUT1 = 'A1 A2 5'
in_list = ['B','L','LB','RB','LT']
```

```python
R : r + 1
L : r - 1
B : c - 1
T : c + 1
RT : r + 1,c + 1
LT : r - 1,c + 1
RB : r + 1,c - 1
LB : r - 1,c - 1
```

```python
King, Rock, N = INPUT1.split()
N = int(N)
```

```python

lo_dict = {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8}
K_r = lo_dict[King[0]]
K_c = int(King[1])
R_r = lo_dict[Rock[0]]
R_c = int(Rock[1])

for i in range(N):
  move = in_list[i]
  if move == "R":
    # 킹 확인
    if K_r + 1 > 8:
      continue
    # 돌 확인
    if (K_r + 1) == R_r and K_c == R_c:
      if R_r + 1 > 8:
        continue
      R_r += 1
    K_r += 1
    
  elif move == "L":
    # 킹 확인
    if K_r - 1 < 1:
      continue
    # 돌 확인
    if (K_r - 1) == R_r and K_c == R_c:
      if R_r - 1 < 1:
        continue
      R_r -= 1
    K_r -= 1

  elif move == "B":
    # 킹 확인
    if K_c - 1 < 1:
      continue
    # 돌 확인
    if K_r == R_r and (K_c - 1) == R_c:
      if R_c - 1 < 1:
        continue
      R_c -= 1
    K_c -= 1

  elif move == "T":
    # 킹 확인
    if K_c + 1 > 8:
      continue
    # 돌 확인
    if K_r == R_r and (K_c + 1) == R_c:
      if R_c + 1 > 8:
        continue
      R_c += 1
    K_c += 1

  elif move == "RT":
    # 킹 확인
    if K_r + 1 > 8 or K_c + 1 > 8:
      continue
    # 돌 확인
    if (K_r + 1) == R_r and (K_c + 1) == R_c:
      if R_r + 1 > 8 or R_c + 1 > 8:
        continue
      R_r += 1
      R_c += 1
    K_r += 1
    K_c += 1

  elif move == "LT":
    # 킹 확인
    if K_r - 1 < 1 or K_c + 1 > 8:
      continue
    # 돌 확인
    if (K_r - 1) == R_r and (K_c + 1) == R_c:
      if R_r - 1 < 1 or R_c + 1 > 8:
        continue
      R_r -= 1
      R_c += 1
    K_r -= 1
    K_c += 1

  elif move == "RB":
    # 킹 확인
    if K_r + 1 > 8 or K_c - 1 < 1:
      continue
    # 돌 확인
    if (K_r + 1) == R_r and (K_c - 1) == R_c:
      if R_r + 1 > 8 or R_c - 1 < 1:
        continue
      R_r += 1
      R_c -= 1
    K_r += 1
    K_c -= 1
    
  elif move == "LB":
    # 킹 확인
    if K_r - 1 < 1 or K_c - 1 < 1:
      continue
    # 돌 확인
    if (K_r - 1) == R_r and (K_c - 1) == R_c:
      if R_r - 1 < 1 or R_c - 1 < 1:
        continue
      R_r -= 1
      R_c -= 1
    K_r -= 1
    K_c -= 1

print(list(lo_dict.keys())[K_r-1]+str(K_c))
print(list(lo_dict.keys())[R_r-1]+str(R_c))
```
